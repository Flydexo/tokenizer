```python
with open("tinychen.txt", "r") as f:
    tinychen = f.read()
```

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0)+1
    return counts
```

```python
data = list(tinychen.encode('utf-8'))
```

```python
def replace_merge(ids, pair, idx):
    i = 0
    new_ids = []
    while i < len(ids):
        if i < len(ids)-1 and (ids[i],ids[i+1]) == pair:
            new_ids.append(idx)
            i+=2
        else:
            new_ids.append(ids[i])
            i+=1
    return new_ids
```

```python
max_tokens = 1000
```

```python
def bytepairencoding(data, tokens={}):
    stats = get_stats(data)
    print(len(tokens), max_tokens)
    if len(tokens) >= max_tokens:
        return data,tokens
    top_pair = max(stats, key=stats.get)
    tokens[top_pair] = len(tokens)+256
    print(f"merging {top_pair} into {tokens[top_pair]}")
    return bytepairencoding(replace_merge(data, top_pair, tokens[top_pair]),tokens)
```

```python
encoded, tokens = bytepairencoding(data)
len(data), len(encoded)
```

```python
def decode(ids, tokens):
    rev_tokens = {v:k for k,v in tokens.items()}
    new_ids = []
    for i in ids:
        new_ids = new_ids + list(rev_tokens.get(i, (i,)))
    if len(new_ids) == len(ids):
        return bytes(new_ids).decode("utf-8", errors='replace')
    return decode(new_ids, tokens)
```

```python
print(decode(encoded[:1000], tokens))
```

```python
def encode(text, tokens):
    if isinstance(text, str):
        by = text.encode('utf-8')
    else:
        by = text
    new_ids = []
    i = 0
    while i < len(by):
        if i < len(by)-1 and tokens.get((by[i], by[i+1])) != None:
            new_ids.append(tokens.get((by[i], by[i+1])))
            i+=2
        else:
            new_ids.append(by[i])
            i+=1
    if len(new_ids) == len(by):
        return new_ids
    return encode(new_ids, tokens)
```
